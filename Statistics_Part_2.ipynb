1. What is hypothesis testing in statistics?
* Hypothesis testing in statistics is a method used to determine whether there is enough evidence to reject a null hypothesis in favor of an alternative hypothesis based on sample data.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
*  The null hypothesis is a statement asserting that there is no effect or no difference, while the alternative hypothesis posits that there is an effect or a difference.

3. What is the significance level in hypothesis testing, and why is it important?
* The significance level, often denoted as alpha (α), is the probability of rejecting the null hypothesis when it is actually true, and it is important because it helps control the risk of making a Type I error.

4.  What does a P-value represent in hypothesis testing?
* A P-value represents the probability of obtaining test results at least as extreme as the observed results, assuming that the null hypothesis is true; it helps determine the strength of evidence against the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?
* In hypothesis testing, a P-value is interpreted as follows: if the P-value is less than or equal to the significance level (α), it suggests strong evidence against the null hypothesis, leading to its rejection; if the P-value is greater than α, it indicates insufficient evidence to reject the null hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing?
* Type I error occurs when the null hypothesis is incorrectly rejected when it is true (false positive), while Type II error occurs when the null hypothesis is not rejected when it is false (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
* A one-tailed test evaluates the possibility of an effect in only one direction (either greater than or less than a certain value), while a two-tailed test assesses the possibility of an effect in both directions (either greater than or less than a certain value).

8. What is the Z-test, and when is it used in hypothesis testing?
* The Z-test is a statistical test used to determine whether there is a significant difference between sample and population means or between two sample means when the population variance is known or the sample size is large (typically n > 30), and the data is approximately normally distributed.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
* The Z-score is calculated using the formulaz = (x – μ) / σ, where  X  is the sample mean, μ  is the population mean, and σ is the population standard deviation; it represents the number of standard deviations a data point is from the mean, helping to determine how unusual or extreme the sample mean is under the null hypothesis.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
* The T-distribution is a probability distribution that is used when the sample size is small (typically n < 30) and/or the population standard deviation is unknown; it is more spread out and has heavier tails than the normal distribution, making it more appropriate for estimating population parameters in these situations.

11.  What is the difference between a Z-test and a T-test?
* The main difference between a Z-test and a T-test is that a Z-test is used when the population variance is known and the sample size is large (n > 30), while a T-test is used when the population variance is unknown and/or the sample size is small (n < 30), relying on the T-distribution for more accurate results in these cases.

12. What is the T-test, and how is it used in hypothesis testing?
* The T-test is a statistical test used to determine whether there is a significant difference between the means of two groups (independent or paired samples) or between a sample mean and a known population mean, particularly when the sample size is small and the population standard deviation is unknown; it helps assess the null hypothesis by comparing the calculated T-value to critical values from the T-distribution.

13. What is the relationship between Z-test and T-test in hypothesis testing?
* The relationship between the Z-test and T-test in hypothesis testing is that both are used to compare means and assess hypotheses, but the Z-test is appropriate for larger sample sizes with known population variance, while the T-test is used for smaller sample sizes with unknown population variance; as sample sizes increase, the T-distribution approaches the normal distribution, making the T-test converge to the Z-test.

14. What is a confidence interval, and how is it used to interpret statistical results?
*  A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter with a specified level of confidence (commonly 95% or 99%); it is used to interpret statistical results by providing an estimate of the uncertainty around the sample statistic, indicating the precision of the estimate and the range within which the true value is expected to lie.

15. What is the margin of error, and how does it affect the confidence interval?
* The margin of error is the amount of uncertainty or potential error in a statistical estimate, calculated as the product of the critical value (from the Z or T distribution) and the standard error; it affects the confidence interval by determining its width, with a larger margin of error resulting in a wider interval, indicating less precision in the estimate.

16. How is Bayes' Theorem used in statistics, and what is its significance?
* Bayes' Theorem is used in statistics to update the probability of a hypothesis based on new evidence or data, providing a mathematical framework for incorporating prior knowledge (prior probability) and the likelihood of observed data; its significance lies in its ability to refine predictions and improve decision-making in various fields, including medical diagnosis, machine learning, and risk assessment.

17. What is the Chi-square distribution, and when is it used?
* The Chi-square distribution is a probability distribution that arises in statistics when analyzing categorical data, particularly in tests of independence and goodness-of-fit; it is used when assessing how well observed data fit an expected distribution or when determining if there is a significant association between categorical variables, typically involving large sample sizes.

18. What is the Chi-square goodness of fit test, and how is it applied?
* The Chi-square goodness of fit test is a statistical test used to determine whether the observed frequencies of a categorical variable match the expected frequencies based on a specific hypothesis or distribution. It is applied by calculating the Chi-square statistic using the formula:

χ2 = ∑(Oi – Ei)2/Ei

where O_i is the observed frequency and E_i is the expected frequency for each category. The calculated Chi-square statistic is then compared to a critical value from the Chi-square distribution table, based on the degrees of freedom (number of categories minus one) and the chosen significance level, to determine if there is a significant difference between the observed and expected frequencies.

19. What is the F-distribution, and when is it used in hypothesis testing?
* The F-distribution is a probability distribution used in hypothesis testing to compare variances between two or more groups, commonly applied in ANOVA and regression analysis.

20. What is an ANOVA test, and what are its assumptions?
* ANOVA (Analysis of Variance) is a statistical test used to determine if there are significant differences between the means of three or more groups. Its assumptions include:
(1) the samples are independent
(2) the data in each group are normally distributed
(3) the variances among the groups are equal (homogeneity of variance). If these assumptions are met, ANOVA can effectively assess group differences.

21. What are the different types of ANOVA tests?
* The different types of ANOVA tests include:

(1) One-Way ANOVA: Compares means across one independent variable with two or more levels (groups).
(2) Two-Way ANOVA: Examines the effect of two independent variables on a dependent variable, including interaction effects between the variables.
(3) Repeated Measures ANOVA: Used when the same subjects are measured multiple times under different conditions or over time.
(4) MANOVA (Multivariate Analysis of Variance): Extends ANOVA to multiple dependent variables, assessing the impact of one or more independent variables on several dependent variables simultaneously.

22. What is the F-test, and how does it relate to hypothesis testing?
* The F-test is a statistical test used to compare the variances of two or more groups to determine if they are significantly different from each other; it is commonly employed in the context of ANOVA and regression analysis. In hypothesis testing, the F-test evaluates the null hypothesis that the group variances are equal by calculating the F-statistic, which is the ratio of the variance estimates, and comparing it to a critical value from the F-distribution to decide whether to reject the null hypothesis.


```python
#Question 1
import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value

sample_data = [20, 22, 21, 19, 23, 20, 22, 21, 20, 19]
population_mean = 21
population_std = 2

z_score, p_value = z_test(sample_data, population_mean, population_std)

print(f"Sample Mean: {np.mean(sample_data)}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")
```

```python
#Question 4
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

sample_data = [48, 50, 52, 49, 51, 53, 50, 48, 52, 50]
population_mean = 50
population_std = 5

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha / 2)

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Z-distribution', color='blue')

plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.5, label='Rejection Region')

plt.axvline(z_score, color='green', linestyle='--', label=f'Z-Score: {z_score:.2f}')

plt.title('Two-Tailed Z-Test')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()

if p_value < alpha:
    print("Reject the null hypothesis: Significant difference exists.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
#Question 5
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(population_mean, population_std, sample_size, true_mean, alpha=0.05):

    critical_value = stats.norm.ppf(1 - alpha / 2)

    x = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)
    y = stats.norm.pdf(x, population_mean, population_std)

    rejection_region_left = population_mean - critical_value * (population_std / np.sqrt(sample_size))
    rejection_region_right = population_mean + critical_value * (population_std / np.sqrt(sample_size))

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Z-distribution', color='blue')

    plt.fill_between(x, y, where=(x <= rejection_region_left) | (x >= rejection_region_right), color='red', alpha=0.5, label='Type I Error Region (α)')

    plt.axvline(rejection_region_left, color='orange', linestyle='--', label='Critical Value Left')
    plt.axvline(rejection_region_right, color='orange', linestyle='--', label='Critical Value Right')

    beta = stats.norm.cdf(rejection_region_left, loc=true_mean, scale=population_std / np.sqrt(sample_size)) + \
           (1 - stats.norm.cdf(rejection_region_right, loc=true_mean, scale=population_std / np.sqrt(sample_size)))

    plt.axvline(true_mean, color='green', linestyle='--', label='True Mean')

    plt.fill_between(x, y, where=(x <= rejection_region_left) | (x >= rejection_region_right) & (x > true_mean), color='yellow', alpha=0.5, label='Type II Error Region (β)')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid()
    plt.show()

    print(f"Type I Error Rate (α): {alpha}")
    print(f"Type II Error Rate (β): {beta:.4f}")

population_mean = 50
population_std = 10
sample_size = 30
true_mean = 55
visualize_errors(population_mean, population_std, sample_size, true_mean)
```

```python
#Question 9
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.
    """
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)
    standard_error = sample_std / np.sqrt(sample_size)

    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    margin_of_error = critical_value * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

sample_data = [23, 21, 18, 25, 30, 22, 19, 24, 27, 20]
ci = confidence_interval(sample_data)
print(f"Confidence Interval: {ci[0]:.2f} to {ci[1]:.2f}")
```

```python
#Question 10
import numpy as np
import scipy.stats as stats

def margin_of_error(sample_data, confidence_level=0.95):
    """
    Calculate the margin of error for a given confidence level.
    """
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    standard_error = sample_std / np.sqrt(sample_size)

    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    return critical_value * standard_error

sample_data = [23, 21, 18, 25, 30, 22, 19, 24, 27, 20]
me = margin_of_error(sample_data)
print(f"Margin of Error: {me:.2f}")
```

```python
#Question 11
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

num_responders = 64
num_tested = 100

bayes_df = pd.DataFrame({'proportion': np.arange(0.0, 1.01, 0.01)})

bayes_df['likelihood'] = stats.binom.pmf(num_responders, num_tested, bayes_df['proportion'])

bayes_df['prior'] = 1 / bayes_df.shape[0]

marginal_likelihood = (bayes_df['likelihood'] * bayes_df['prior']).sum()

bayes_df['posterior'] = (bayes_df['likelihood'] * bayes_df['prior']) / marginal_likelihood

plt.plot(bayes_df['proportion'], bayes_df['likelihood'], label='Likelihood')
plt.plot(bayes_df['proportion'], bayes_df['prior'], label='Prior')
plt.plot(bayes_df['proportion'], bayes_df['posterior'], 'k--', label='Posterior')
plt.legend()
plt.xlabel('Proportion of Responders')
plt.ylabel('Probability')
plt.title('Bayesian Inference using Bayes\' Theorem')
plt.show()
```

```python
#Question 13
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate the expected frequencies for a Chi-square test based on observed data.

    Parameters:
    observed (array-like): The observed frequency table.

    Returns:
    np.array: The expected frequency table.
    """

    observed_df = pd.DataFrame(observed)

    row_totals = observed_df.sum(axis=1).values.reshape(-1, 1)
    column_totals = observed_df.sum(axis=0).values
    grand_total = observed_df.sum().sum()

    expected = (row_totals * column_totals) / grand_total

    return expected

observed_data = np.array([[10, 20, 30],
                           [6,  14,  20]])

expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:\n", observed_data)
print("Expected Frequencies:\n", expected_frequencies)
```

```python
#Question 15
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_chi_square_distribution(degrees_of_freedom, x_range):
    """
    Plot the Chi-square distribution for given degrees of freedom.

    Parameters:
    degrees_of_freedom (list): List of degrees of freedom to plot.
    x_range (tuple): Range of x values for the plot.
    """
    x = np.linspace(x_range[0], x_range[1], 1000)

    plt.figure(figsize=(10, 6))

    for df in degrees_of_freedom:

        y = stats.chi2.pdf(x, df)
        plt.plot(x, y, label=f'DF = {df}')

    plt.title('Chi-square Distribution')
    plt.xlabel('Value')
    plt.ylabel('Probability Density Function (PDF)')
    plt.legend()
    plt.grid()
    plt.xlim(x_range)
    plt.ylim(0, 0.5)
    plt.show()

degrees_of_freedom = [1, 2, 5, 10]
x_range = (0, 30)

plot_chi_square_distribution(degrees_of_freedom, x_range)
```

```python
#Question 19
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(groups):
    """
    Check the assumptions for ANOVA: normality and equal variance.

    Parameters:
    groups (list of arrays): List of arrays containing the data for each group.
    """

    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f'Group {i+1} - Shapiro-Wilk: W-statistic = {stat:.4f}, p-value = {p_value:.4f}')

    stat, p_value = stats.levene(*groups)
    print(f'Levene\'s Test: W-statistic = {stat:.4f}, p-value = {p_value:.4f}')

    plt.figure(figsize=(10, 6))
    for i, group in enumerate(groups):
        sns.kdeplot(group, label=f'Group {i+1}', fill=True)
    plt.title('Kernel Density Estimates of Groups')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

group1 = np.array([23, 21, 18, 25, 30])
group2 = np.array([30, 32, 29, 35, 28])
group3 = np.array([22, 24, 19, 23, 21])

check_anova_assumptions([group1, group2, group3])
```

```python
#Question 20
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
data = {
    'Factor_A': np.repeat(['A1', 'A2'], 20),
    'Factor_B': np.tile(np.repeat(['B1', 'B2'], 10), 2),
    'Response': np.random.randn(40) + np.repeat([1, 2, 1.5, 2.5], 10)
}

df = pd.DataFrame(data)

model = ols('Response ~ C(Factor_A) * C(Factor_B)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print(anova_table)

plt.figure(figsize=(10, 6))
sns.pointplot(x='Factor_A', y='Response', hue='Factor_B', data=df,
               markers=['o', 's'], linestyles=['-', '--'], ci=None)
plt.title('Interaction Plot')
plt.xlabel('Factor A')
plt.ylabel('Response')
plt.legend(title='Factor B')
plt.grid()
plt.show()
```

```python
#Question 21
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_f_distribution(degrees_of_freedom):
    """
    Plot the F-distribution for given degrees of freedom.

    Parameters:
    degrees_of_freedom (list): List of tuples containing (df1, df2) for the F-distribution.
    """
    x = np.linspace(0, 5, 1000)

    plt.figure(figsize=(10, 6))

    for df1, df2 in degrees_of_freedom:
        y = stats.f.pdf(x, df1, df2)
        plt.plot(x, y, label=f'df1={df1}, df2={df2}')

    plt.title('F-Distribution')
    plt.xlabel('F-value')
    plt.ylabel('Probability Density Function (PDF)')
    plt.legend()
    plt.grid()
    plt.xlim(0, 5)
    plt.ylim(0, 1)
    plt.show()

degrees_of_freedom = [(1, 1), (5, 2), (10, 10), (5, 5)]
plot_f_distribution(degrees_of_freedom)
```

```python
#Question 22
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(0)
group1 = np.random.normal(loc=20, scale=5, size=30)
group2 = np.random.normal(loc=22, scale=5, size=30)
group3 = np.random.normal(loc=25, scale=5, size=30)

data = pd.DataFrame({
    'Group': np.repeat(['Group 1', 'Group 2', 'Group 3'], 30),
    'Value': np.concatenate([group1, group2, group3])
})

f_statistic, p_value = stats.f_oneway(group1, group2, group3)

print(f"F-Statistic: {f_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: At least one group mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between group means.")

plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Value', data=data)
plt.title('Boxplot of Group Values')
plt.xlabel('Group')
plt.ylabel('Value')
plt.grid()
plt.show()
```

```python
#Question 25
import numpy as np
import scipy.stats as stats

def z_test_proportions(success_a, n_a, success_b, n_b, alpha=0.05):
    """
    Perform a Z-test for comparing proportions between two groups.

    Parameters:
    - success_a: Number of successes in group A
    - n_a: Total number of trials in group A
    - success_b: Number of successes in group B
    - n_b: Total number of trials in group B
    - alpha: Significance level (default is 0.05)

    Returns:
    - Z-statistic and p-value
    """
    p_a = success_a / n_a
    p_b = success_b / n_b

    p_pool = (success_a + success_b) / (n_a + n_b)

    z_statistic = (p_a - p_b) / np.sqrt(p_pool * (1 - p_pool) * (1/n_a + 1/n_b))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    return z_statistic, p_value

success_a = 30
n_a = 100
success_b = 20
n_b = 80

z_statistic, p_value = z_test_proportions(success_a, n_a, success_b, n_b)

print(f"Z-Statistic: {z_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```

```python
#Question 26
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=55, scale=15, size=100)

f_statistic, p_value = stats.levene(data1, data2)

print(f"F-Statistic: {f_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

plt.figure(figsize=(10, 6))
sns.boxplot(data=[data1, data2], palette='Set2')
plt.title('Boxplot of Two Datasets')
plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])
plt.ylabel('Values')
plt.grid()
plt.show()
```

```python
#Question 27
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(0)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=60)
observed_counts = np.array([np.sum(observed == i) for i in range(1, 7)])
expected_counts = np.array([10] * 6)

chi2_statistic, p_value = stats.chisquare(observed_counts, expected_counts)

print(f"Chi-square Statistic: {chi2_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")

categories = np.arange(1, 7)
plt.bar(categories - 0.2, observed_counts, width=0.4, label='Observed', color='blue')
plt.bar(categories + 0.2, expected_counts, width=0.4, label='Expected', color='orange')
plt.xlabel('Die Faces')
plt.ylabel('Counts')
plt.title('Chi-square Goodness of Fit Test')
plt.xticks(categories)
plt.legend()
plt.show()
```
